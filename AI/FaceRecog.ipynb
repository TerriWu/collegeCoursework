```python
import cv2
import sys
import gc
from google.colab.patches import cv2_imshow
from FaceTrain_Niaws import Model
```

```python
#載入模型
model = Model()
model.load_model(file_path = "./model/Niaws_model.h5")
              
#框住臉部的矩形顏色       
color = (255, 0, 255)
    
#捕獲指定的錄影檔案
cap = cv2.VideoCapture('./Niaws_video.mp4')
    
#人臉識別分類器本地儲存路徑
cascade_path = "./haarcascade_frontalface_alt2.xml"    
    
#迴圈檢測識別人臉
while True:
  ret, frame = cap.read()   #讀取一Frame

  if ret is True:
    #GrayScale降低計算時的複雜度
    frame_gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
  else:
    continue

  #使用OpenCV人臉分類器
  cascade = cv2.CascadeClassifier(cascade_path)                

  #先找出影像中哪個區域為人臉
  faceRects = cascade.detectMultiScale(frame_gray, scaleFactor = 1.2, minNeighbors = 3, minSize = (32, 32))        
  if len(faceRects) > 0:
    for faceRect in faceRects: 
      x, y, w, h = faceRect
                
      #再把臉部影象擷取後，交給model看是誰
      image = frame[y - 10: y + h + 10, x - 10: x + w + 10]
      faceID = model.face_predict(image)   

      #如果就是我Laisan
      if faceID == 0:
        cv2.rectangle(frame, (x - 10, y - 10), (x + w + 10, y + h + 10), color, thickness = 2)
        
        #文字提示
        cv2.putText(frame,'Niaws', 
                  (x + 30, y + 30),               #座標
                  cv2.FONT_HERSHEY_SIMPLEX,              #字型
                  1,                         #字號
                  (255,0,255),                    #顏色
              2)                             #字的線寬

        cv2_imshow(frame)
      else:
        pass
                            
      

#釋放Camera並關閉視窗
cap.release()
cv2.destroyAllWindows()
```
