```python
import cv2
import sys
from PIL import Image
from google.colab.patches import cv2_imshow  #使用google colab 需打這行
```

```python
def CatchVideo(window_name, camera_idx):
    cv2.namedWindow(window_name)
    
    #影像的來源  camera_idx是鏡頭代碼 (可寫影片路徑偵測影片中是否是人臉)
    #使用人臉的鼻子、眼睛、嘴巴…等等，比例
    #戴面罩、口罩偵測不到是因為沒有定義戴面罩、口罩(比例不同) 
    #cap = cv2.VideoCapture(0) 0==電腦預設的第一個鏡頭 
    cap = cv2.VideoCapture("./59406972.mp4")                
    
    #告訴OpenCV使用人臉分類器
    classfier = cv2.CascadeClassifier("./haarcascade_frontalface_alt2.xml")
    
    #識別後要畫的框用色
    color = (255, 0, 255)
        
    while cap.isOpened():
        ok, frame = cap.read() #讀取一frame
        if not ok:            
            break  

        #轉成GrayScale以減少記憶量 圖片變成灰階
        grey = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)                 
        
        #人臉檢測，1.2和2分別為圖片縮放比例和需要檢測的有效點數
        faceRects = classfier.detectMultiScale(grey, scaleFactor = 1.2, minNeighbors = 3, minSize = (32, 32))
        if len(faceRects) > 0:          #偵測到多少人臉                                   
            for faceRect in faceRects:  #框出每一張人臉
                x, y, w, h = faceRect        
                cv2.rectangle(frame, (x - 10, y - 10), (x + w + 10, y + h + 10), color, 2)
                        
        #顯示處理後的  cv2_imshowl：google colab寫法 cv2imshow：正常寫法
      
        cv2_imshow(window_name, frame)        
        c = cv2.waitKey(10)
        if c & 0xFF == ord('q'):
            break        
    
    #釋放Camera並關閉視窗
    cap.release()
    cv2.destroyAllWindows() 
  
```

```python
if __name__ == '__main__':
    if len(sys.argv) != 1:
        print("Usage:%s camera_id\r\n" % (sys.argv[0]))
    else:
        CatchVideo("Face detection", 0)
```
